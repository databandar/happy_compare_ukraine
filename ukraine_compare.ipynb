```python
df['country'] = df['country'].replace({
     "Taiwan Province of China": "Taiwan",
    "Slovakia": "Slovak Republic",
    "Iran": "Islamic Republic of Iran",
    "Egypt": "Arab Republic of Egypt",
    "Republic of Moldova": "Moldova",
    "Czechia": "Czech Republic",
    "Hong Kong SAR of China": "Hong Kong, SAR",
    "Lao PDR": "Lao People's Democratic Republic",
    "Kyrgyzstan": "Kyrgyz Republic",
    "Yemen": "Republic of Yemen",
    "Côte d’Ivoire": "Côte d'Ivoire",  # smart quote to straight quote
    "Türkiye": "Turkey",
    "DR Congo": "Democratic Republic of Congo",
    "Venezuela": "R. B. de Venezuela",
    "Viet Nam": "Vietnam",
    "North Macedonia": "FYR of Macedonia",
    "State of Palestine": "West Bank and Gaza",
    "United States": "United States of America",
    "Gambia": "The Gambia"
})
```

```python
merged_gdf = gdf.merge(df, how='left', left_on='name', right_on='country')
```

```python
merged_gdf['score'] = merged_gdf['score'].fillna(0).astype(float)
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import matplotlib.colors as mcolors
import numpy as np
import pandas as pd
from matplotlib.cm import ScalarMappable
from matplotlib.colorbar import ColorbarBase
from matplotlib.colors import ListedColormap

# Ensure score is numeric
merged_gdf['score'] = pd.to_numeric(merged_gdf['score'], errors='coerce')

# Get Ukraine's score
ukraine_score = merged_gdf.loc[merged_gdf['name'] == 'Ukraine', 'score'].values[0]

# Categorize for binning
def categorize(row):
    if pd.isna(row['score']) or row['score'] == 0:
        return 'No Data'
    elif row['name'] == 'Ukraine':
        return 'Ukraine'
    elif row['score'] > ukraine_score:
        return 'Happier'
    else:
        return 'Sadder'

merged_gdf['category'] = merged_gdf.apply(categorize, axis=1)

# Define bins
happier_bins = np.linspace(ukraine_score, merged_gdf['score'].max(), 6)
sadder_bins = np.linspace(merged_gdf['score'].min(), ukraine_score, 6)

# Custom colormaps
happier_cmap = ListedColormap(plt.cm.Blues(np.linspace(0.4, 0.9, 256)))

# Custom sadder green colormap (dark to medium green, avoiding pale tones)
custom_green_colors = [
    '#00441b',  # very dark green
    '#006d2c',
    '#238b45',
    '#41ab5d',
    '#74c476',
    '#a1d99b'   # avoid anything lighter
]
sadder_cmap = ListedColormap(custom_green_colors)

# Normalize
happier_norm = mcolors.BoundaryNorm(happier_bins, happier_cmap.N)
sadder_norm = mcolors.BoundaryNorm(sadder_bins, len(custom_green_colors))

# Assign colors
def get_color(row):
    if row['category'] == 'No Data':
        return '#555555'  # dark grey
    elif row['category'] == 'Ukraine':
        return '#FFD700'  # gold
    elif row['category'] == 'Happier':
        return happier_cmap(happier_norm(row['score']))
    elif row['category'] == 'Sadder':
        return sadder_cmap(sadder_norm(row['score']))
    else:
        return '#ffffff'

merged_gdf['plot_color'] = merged_gdf.apply(get_color, axis=1)

# Plot
plt.style.use('default')
fig, ax = plt.subplots(1, 1, figsize=(20, 25))
fig.patch.set_facecolor('#fffde7') 

merged_gdf.plot(ax=ax,
                color=merged_gdf['plot_color'],
                edgecolor='white',
                linewidth=0.5,
                alpha=0.9)

# Title
ax.text(0.01, 0.45, 
        "Is Your Country Happier \n or Sadder Than Ukraine?",
        transform=ax.transAxes,
        fontsize=24, fontweight='bold',
        color='#2c3e50', fontfamily='serif',
        ha='left', va='top')

# Axis limits and background
ax.set_xlim(-180, 180)
ax.set_ylim(-60, 85)
ax.axis('off')
ax.set_facecolor('#f0f8ff')

# Colorbars (horizontal)
cbar_width = 0.25
cbar_height = 0.015
cbar_y1 = 0.31
cbar_y2 = 0.38
legend_y = 0.13  # below colorbars

# Sadder (left, reversed)
cbar_ax1 = fig.add_axes([0.01, cbar_y1, cbar_width, cbar_height])
sadder_sm = ScalarMappable(norm=sadder_norm, cmap=sadder_cmap)
sadder_cbar = ColorbarBase(cbar_ax1, cmap=sadder_cmap, norm=sadder_norm, orientation='horizontal')
sadder_cbar.set_label('Sadder Scores', fontsize=11, labelpad=5,fontweight='bold')
sadder_cbar.ax.tick_params(labelsize=9)
sadder_cbar.ax.invert_xaxis()  # reverse direction

# Happier (right)
cbar_ax2 = fig.add_axes([0.01, cbar_y2, cbar_width, cbar_height])
happier_sm = ScalarMappable(norm=happier_norm, cmap=happier_cmap)
happier_cbar = ColorbarBase(cbar_ax2, cmap=happier_cmap, norm=happier_norm, orientation='horizontal')
happier_cbar.set_label('Happier Scores', fontsize=11, labelpad=5,fontweight='bold')
happier_cbar.ax.tick_params(labelsize=9)

# Horizontal legend below colorbars
legend_elements = [
    plt.Rectangle((0, 0), 1, 1, facecolor='#FFD700', edgecolor='white', label='Ukraine (4.68)'),
    plt.Rectangle((0, 0), 1, 1, facecolor='#555555', edgecolor='white', label='No Data'),
]
legend = ax.legend(handles=legend_elements,
                   loc='lower left',
                   bbox_to_anchor=(0.02, legend_y),
                   ncol=2,
                   frameon=True,
                   fancybox=True,
                   shadow=True,
                   fontsize=14,
                   handlelength=1.5)
legend.get_frame().set_facecolor('#fffde7')
legend.get_frame().set_alpha(0.95)
legend.get_frame().set_edgecolor('#bdc3c7')
legend.get_frame().set_linewidth(1)

# Methodology box
methodology_text = (
    "Methodology:\n"
    "The Gallup World Poll asks respondents to rate their life satisfaction\n"
    "on a scale from 0 (worst possible life) to 10 (best possible life).\n\n"
    "Country scores represent the average response from ~1,000 adults,\n"
    "calculated as three-year rolling averages for improved reliability."
)
textbox_props = dict(boxstyle="round,pad=0.8", facecolor='#fffde7', 
                     alpha=0.9, edgecolor='#CCCCCC', linewidth=1)
ax.text(0.35, 0.08, methodology_text,
        transform=ax.transAxes,
        fontsize=12, 
        fontweight='bold',
        verticalalignment='top',
        horizontalalignment='left',
        color='#2C3E50',
        bbox=textbox_props,
        linespacing=1.4)

# Source
ax.text(0.98, 0.02, 
        'Data Source: World Happiness Report 2025',
        transform=ax.transAxes, 
        fontsize=10, 
        style='italic',
        verticalalignment='bottom',
        horizontalalignment='right',
        color='#7F8C8D')

plt.tight_layout()
plt.subplots_adjust(top=0.85, bottom=0.05, left=0.02, right=0.98)
plt.savefig('happiness2.png', dpi=300, bbox_inches='tight', facecolor=fig.get_facecolor())

plt.show()
```
