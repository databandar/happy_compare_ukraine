```python
import pandas as pd

# Step 1: Drop the 'gdp_per_capita_growth' column
df = df.drop(columns=['gdp_per_capita_growth'])

# Step 2: Pivot the data to wide format with years as columns
df_pivot = df.pivot(index='country', columns='date', values='gdp_per_capita')

# Step 3: Calculate % change from 2000 to 2024
df_pivot['gdp_growth_2014_2024'] = ((df_pivot['2024'] - df_pivot['2014']) / df_pivot['2014']) * 100

# Step 4: Reset index (optional)
df_growth = df_pivot[['gdp_growth_2014_2024']].reset_index()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.cm import ScalarMappable
import matplotlib.patches as mpatches
import matplotlib as mpl
import pandas as pd
import numpy as np

# Set global font to serif and bold
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.weight'] = 'bold'

# Ensure GDP column is numeric
merged_gdf['gdp_growth_2014_2024'] = pd.to_numeric(merged_gdf['gdp_growth_2014_2024'], errors='coerce')

# Get Guyana's score for reference
guyana_score = merged_gdf.loc[merged_gdf['country'] == 'Guyana', 'gdp_growth_2014_2024'].values[0]

# Define bins and custom color palette
bins = [-60, -40, -20, -10, 0, 10, 20, 40, 60, 80, 120]
colors = [
   '#67000d', '#a50f15', '#cb181d', '#ef3b2c',  # Reds to yellow (negative growth)
    '#e0f3f8', '#abd9e9', '#74add1', '#4575b4', '#313695', '#1a1a5e'  # Light to dark blues (positive growth)
]
cmap = mcolors.ListedColormap(colors)
norm = mcolors.BoundaryNorm(bins, len(colors))

# Create figure with light yellow background
plt.style.use('default')
fig, ax = plt.subplots(1, 1, figsize=(20, 25))
fig.patch.set_facecolor('#fffde7') 

# Plot the map with white borders
merged_gdf.plot(
    column='gdp_growth_2014_2024',
    cmap=cmap,
    norm=norm,
    linewidth=0.5,
    edgecolor='white',
    ax=ax,
    alpha=0.9,
    missing_kwds={'color': '#555555', 'edgecolor': 'white'}
)

# Highlight Guyana in gold with white border
guyana = merged_gdf[merged_gdf['country'] == 'Guyana']
guyana.plot(color='#FFD700', edgecolor='white', linewidth=0.5, ax=ax, alpha=0.9)

# Title with reduced font size
ax.text(-0.01, 0.45, 
        "GDP per Capita Growth \n (2014â€“2024)",
        transform=ax.transAxes,
        fontsize=24,
        fontweight='bold',
        color='#2c3e50', 
        fontfamily='serif',
        ha='left', 
        va='top')

# Title with reduced font size
ax.text(-0.01, 0.30, 
        "Percentage Change in GDP per capita from 2014 \n to 2024 (PPP-adjusted, constant 2021 dollars)",
        transform=ax.transAxes,
        fontsize=14,
        fontweight='bold',
        color='#2c3e50', 
        fontfamily='serif',
        ha='left', 
        va='top')

# Axis limits and background
ax.set_xlim(-180, 180)
ax.set_ylim(-60, 85)
ax.axis('off')
ax.set_facecolor('#f0f8ff')

# Create horizontal colorbar moved up
cbar_width = 0.4
cbar_height = 0.02
cbar_y = 0.27

cbar_ax = fig.add_axes([0.3, cbar_y, cbar_width, cbar_height])
sm = ScalarMappable(cmap=cmap, norm=norm)
cbar = fig.colorbar(
    sm,
    cax=cbar_ax,
    orientation='horizontal',
    ticks=bins
)
cbar.set_label('GDP per capita PPP Growth (%)', fontsize=12, labelpad=10, fontweight='bold')
cbar.ax.tick_params(labelsize=10)

# Legend elements moved next to colorbar
legend_elements = [
    mpatches.Patch(facecolor='#FFD700', edgecolor='white', label=f'Guyana ({guyana_score:.1f}%)'),
    mpatches.Patch(facecolor='#555555', edgecolor='white', label='No Data'),
]

# Place legend to the right of the colorbar
legend = ax.legend(handles=legend_elements,
                   loc='center',
                   bbox_to_anchor=(0.22, -0.05),
                   frameon=True,
                   fancybox=True,
                   shadow=True,
                   fontsize=12,
                   handlelength=1.5)
legend.get_frame().set_facecolor('#fffde7')
legend.get_frame().set_alpha(0.95)
legend.get_frame().set_edgecolor('#bdc3c7')
legend.get_frame().set_linewidth(1)



# Source
ax.text(0.98, -0.05,
        'Data Source: World Bank (2024) \n Created by : @DataBandar',
        transform=ax.transAxes, 
        fontsize=11, 
        
        verticalalignment='bottom',
        horizontalalignment='right',
        color="#4A2A05")

plt.tight_layout()
plt.subplots_adjust(top=0.85, bottom=0.05, left=0.02, right=0.98)
plt.savefig('gdp_capitaone.png', dpi=300, bbox_inches='tight', facecolor=fig.get_facecolor())
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.cm import ScalarMappable
import matplotlib.patches as mpatches
import matplotlib as mpl
import pandas as pd
import numpy as np

# Set global font to serif and bold
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.weight'] = 'bold'

# Ensure GDP column is numeric
merged_gdf['gdp_growth_2014_2024'] = pd.to_numeric(merged_gdf['gdp_growth_2014_2024'], errors='coerce')

# Get Guyana's score for reference
guyana_score = merged_gdf.loc[merged_gdf['country'] == 'Guyana', 'gdp_growth_2014_2024'].values[0]

# Define bins and custom color palette
bins = [-60, -40, -20, -10, 0, 10, 20, 40, 60, 80, 120]
colors = [
   '#67000d', '#a50f15', '#cb181d', '#ef3b2c',  # Reds to yellow (negative growth)
    '#e0f3f8', '#abd9e9', '#74add1', '#4575b4', '#313695', '#1a1a5e'  # Light to dark blues (positive growth)
]
cmap = mcolors.ListedColormap(colors)
norm = mcolors.BoundaryNorm(bins, len(colors))

# Create figure with light yellow background
plt.style.use('default')
fig, ax = plt.subplots(1, 1, figsize=(20, 12))  # Reduced height for better proportions
fig.patch.set_facecolor('#fffde7')

# Plot the map with white borders
merged_gdf.plot(
    column='gdp_growth_2014_2024',
    cmap=cmap,
    norm=norm,
    linewidth=0.5,
    edgecolor='white',
    ax=ax,
    alpha=0.9,
    missing_kwds={'color': '#555555', 'edgecolor': 'white'}
)

# Highlight Guyana in gold with white border
guyana = merged_gdf[merged_gdf['country'] == 'Guyana']
guyana.plot(color='#FFD700', edgecolor='white', linewidth=0.5, ax=ax, alpha=0.9)

# Set main title - positioned at the top left
fig.text(0.05, 0.95, 
         "Global GDP per Capita Growth (2014â€“2024): Guyana Leads with 484.3% Increase",
         fontsize=24,
         fontweight='bold',
         color='#2c3e50',
         fontfamily='serif',
         verticalalignment='top')

# Set subtitle - positioned below main title
fig.text(0.05, 0.91, 
         "Percentage Change in GDP per capita from 2014 to 2024 (PPP-adjusted, constant 2021 dollars)",
         fontsize=16,
         fontweight='normal',
         color='#2c3e50',
         fontfamily='serif',
         verticalalignment='top')

# Axis limits and background
ax.set_xlim(-180, 180)
ax.set_ylim(-60, 85)
ax.axis('off')
ax.set_facecolor('#f0f8ff')

# Create horizontal colorbar - positioned in upper middle area
cbar_width = 0.35
cbar_height = 0.02
cbar_x = 0.325  # Centered horizontally
cbar_y = 0.82   # Below the titles
cbar_ax = fig.add_axes([cbar_x, cbar_y, cbar_width, cbar_height])

sm = ScalarMappable(cmap=cmap, norm=norm)
cbar = fig.colorbar(
    sm,
    cax=cbar_ax,
    orientation='horizontal',
    ticks=bins
)
cbar.set_label('GDP per capita PPP Growth (%)', fontsize=12, labelpad=10, fontweight='bold')
cbar.ax.tick_params(labelsize=10)
# Set colorbar tick labels to serif and bold
for label in cbar.ax.get_xticklabels():
    label.set_fontfamily('serif')
    label.set_fontweight('bold')

# Annotate Guyana directly on the map
guyana_centroid = guyana.geometry.centroid.iloc[0]
guyana_x = guyana_centroid.x
guyana_y = guyana_centroid.y

# Add simple annotation with arrow pointing to Guyana
ax.annotate('Guyana', 
            xy=(guyana_x, guyana_y),
            xytext=(guyana_x + 15, guyana_y + 10),  # Offset to the left
            fontsize=11,
            fontweight='bold',
            color='#2c3e50',
            ha='center',
            va='bottom',
            arrowprops=dict(arrowstyle='->', 
                          color='#2c3e50', 
                          lw=2,
                          connectionstyle="arc3,rad=0.3"))  # This creates a curvy arrow

# Simplified legend with only "No Data"
legend_elements = [
    mpatches.Patch(facecolor='#555555', edgecolor='white', label='No Data'),
]

# Create simplified legend positioned to the right of the colorbar
legend_x = cbar_x + cbar_width + 0.05  # To the right of colorbar
legend_y = cbar_y + (cbar_height / 2)  # Vertically centered with colorbar

legend = fig.legend(handles=legend_elements,
                   loc='center',
                   bbox_to_anchor=(legend_x + 0.05, legend_y),
                   frameon=True,
                   fancybox=True,
                   shadow=True,
                   fontsize=12,
                   handlelength=1.5)

legend.get_frame().set_facecolor('#fffde7')
legend.get_frame().set_alpha(0.95)
legend.get_frame().set_edgecolor('#bdc3c7')
legend.get_frame().set_linewidth(1)

# Source - positioned at bottom right
fig.text(0.18, 0.12,
         'Data Source: World Bank (2024)\nCreated by: @DataBandar',
         fontsize=11,
         verticalalignment='bottom',
         horizontalalignment='right',
         color="#1C160F")

# Adjust layout to accommodate titles and elements
plt.subplots_adjust(top=0.78, bottom=0.05, left=0.02, right=0.98)
plt.savefig('gdp_capita.png', dpi=300, bbox_inches='tight', facecolor=fig.get_facecolor())
plt.show()
```
